### Bootstrap

```python
results = []

for nrepreat in range(1000):
    sample = resample(loans_income)
    results.append(sample.median())
    
results = pd.Series(results)
print('Bootstrap statistics: ')
print(f'\noriginal: {loans_income.median()}')
print(f'\nbias: {results.mean() - loans_income.median()}')
print(f'standard deviation: {results.std()}')
```

```python
# Generate example data (exam scores)
np.random.seed(42)  # for reproducibility
exam_scores = np.random.normal(loc=70, scale=10, size=100)  # mean=70, std=10, 100 samples

def bootstrap_mean(data, n_bootstrap=1000, confidence_level=0.95):
    bootstrap_means = []
    n = len(data)
    
    for _ in range(n_bootstrap):
        # Resample with replacement
        bootstrap_sample = np.random.choice(data, size=n, replace=True)
        # Calculate mean of bootstrap sample
        bootstrap_mean = np.mean(bootstrap_sample)
        bootstrap_means.append(bootstrap_mean)
    
    # Calculate confidence interval
    alpha = (1 - confidence_level) / 2
    lower_bound = np.percentile(bootstrap_means, alpha * 100)
    upper_bound = np.percentile(bootstrap_means, (1 - alpha) * 100)
    
    return np.mean(bootstrap_means), (lower_bound, upper_bound)

# Apply bootstrapping to estimate mean and confidence interval
mean_estimate, confidence_interval = bootstrap_mean(exam_scores)

print("Estimated mean:", mean_estimate)
print("95% Confidence Interval:", confidence_interval)
```
