### 1. Example: Location estimates of Population and Murder rates

Calculate mean, trimmed mean and median for the population

```python
import wquantiles as wq
import numpy as np
import pandas as pd
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('practical-statistics-for-data-scientists-master\data\state.csv')
df.head()
```

```python
df.shape
```

```python
mean = df['Population'].mean()
trim_mean = scipy.stats.trim_mean(df['Population'], 0.1)
median = df['Population'].median()

print('The median is: ', median)
print('The mean is: ', mean)
print('The trimmed mean is: ', trim_mean)

print('\nmean/trim_mean = ', mean/trim_mean)
print('mean/median = ', mean/median)
print('trim_mean/median = ', trim_mean/median)
```

Isso aqui abaixo eu fiz pra verificar que, pra calcular trimmed mean, np primeiro ordena o arranjo, e depois calcula o mean dentro do intervalo de interesse

```python
pop = df['Population'].sort_values()[5:45]
```

```python
pop.mean()
```

To calculate the average murder rate for the country, we can calculate the **weighted mean** or **weighted median**.

```python
weighted_mean = np.average(df['Murder.Rate'], weights=df['Population'])
weigthed_median = wq.median(df['Murder.Rate'], weights=df['Population']) 

print('weighted mean = ', weighted_mean)
print('weigthed median = ', weigthed_median)
```

### Percentiles and boxplots (whiskers)

```python
df.columns
```

```python
Q1 = df['Population'].quantile(0.25)
Q2 = df['Population'].quantile(0.50)
Q3 = df['Population'].quantile(0.75)

IQR = Q3 - Q1

w_max = Q3 + 1.5 * IQR
w_min = Q1 - 1.5 * IQR

# Find the minimum and maximum values within the whiskers' range
min_whisker = df['Population'].quantile(0.25) - 1.5 * IQR
max_whisker = df['Population'].quantile(0.75) + 1.5 * IQR
```

```python
(df['Population']/1e6).plot.box()

plt.scatter(np.ones(len(df['Population'])), df['Population']/1e6, c='white', edgecolor='k')

plt.axhline(y=Q1/1e6, c='tab:blue', label = 'Q1')
plt.axhline(y=Q2/1e6, c='tab:red', label = 'Q2')
plt.axhline(y=Q3/1e6, c='tab:green', label = 'Q3')

# plt.axhline(y=IQR/1e6, c='tab:orange', label = 'IQR')

plt.axhline(y= max_whisker /1e6 , c='k', ls='-', label = 'w_max')
plt.axhline(y= min_whisker /1e6 , c='k', ls='--', label = 'w_min')

plt.legend()
plt.show()
```

### Hexagonal binning and contour plots

Let us use the data set containing the tax-assessed values for residential properties in King county, Washington. 

```python
path = r'C:\Users\josea\OneDrive\Desktop\Documents\Statistics\practical-statistics-for-data-scientists-master\data\kc_tax_data.csv'
kc_tax = pd.read_csv(path)

kc_tax0 = kc_tax.loc[(kc_tax.TaxAssessedValue < 750000) &
                     (kc_tax.SqFtTotLiving > 100) &
                     (kc_tax.SqFtTotLiving < 3500), :]

kc_tax0.shape
```

### Boxplots and violin plots for comparison of numerical and categorical data

Let us compare how the percentage of flight delays varies across airlines.

```python
filepath = r'C:\Users\josea\OneDrive\Desktop\Documents\Statistics\practical-statistics-for-data-scientists-master\data\airline_stats.csv'

airline_stats = pd.read_csv(filepath)

airline_stats.head()
```

```python
airline_stats.shape
```

```python
airline_stats.boxplot(by='airline', column='pct_carrier_delay')
plt.xlabel('')
plt.ylabel('Daily percentage of delayed flights')
plt.title('')
plt.suptitle('')
plt.ylim(top=52)
plt.show()
```

```python
sns.violinplot(airline_stats.airline, airline_stats.pct_carrier_delay, inner='quartile', zorder=5)
plt.xlabel('')
plt.ylabel('Daily percentage of delayed flights')
plt.title('')
plt.suptitle('')
plt.ylim(top=52)
plt.show()
```
